```python
#this is the not as clean way to get the mean with sorted values, but it still works
mean= uni.groupby(['Type'])['Size'].mean().sort_values()
mean
```

```python
#here i've set median equal to a variable so I can call it elsewhere in short hand
med= uni.groupby(['Type'])['Size'].median()
med

```

```python
#I call it here in shorthand and sort the values in a clean way
med=med.sort_values()
med
```

```python
address1 = uni['address']
address1 = address1.str.strip()
address1 = address1.str.split(',', expand=True) #expand puts results into two separate columns
address2 = address1.rename(columns={0:'street', 1:'city', 2:'state zip'})
address2
```

```python
address3 = address2['state zip'].str.rsplit(' ',1, expand=True)
address3= address3.rename(columns={0:'state', 1: 'zip'})
address3
```

```python
data1= pd.concat([uni,address2,address3],axis=1)
data1
```

```python
data1 = data1.drop(columns = ['address','state zip', 3, 4, 5])
data1
```
